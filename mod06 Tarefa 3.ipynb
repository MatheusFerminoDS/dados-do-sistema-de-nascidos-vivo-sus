#### 1) A base 'sinasc_RO_2019.csv' contém dados do sistema de nascidos vivos do DataSUS, cujo dicionário de dados está no arquivo 'Estrutura_SINASC_para_CD.pdf'. Carregue esse arquivo.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('sinasc_RO_2019.csv')

df.head()
```

#### 2) Categorize a variável APGAR5 conforme as seguintes quebras:

- 0 a 3: 'asfixia grave'
- 4 a 5: 'asfixia moderada'
- 6 a 7: 'asfixia leve'
- 8 a 10: 'normal'

```python
def categorizar_apgar5(valor):
    if valor >= 0 and valor <= 3:
        return 'asfixia grave'
    elif valor >= 4 and valor <= 5:
        return 'asfixia moderada'
    elif valor >= 6 and valor <= 7:
        return 'asfixia leve'
    elif valor >= 8 and valor <= 10:
        return 'normal'

#riar uma nova coluna categorizada
df['APGAR5_cat'] = df['APGAR5'].apply(categorizar_apgar5)

df[['APGAR5', 'APGAR5_cat']].head()
```

#### 3) Monte um gráfico comparando a distribuição de APGAR5 por sexo para cada grupo de tempo de gestação (GESTACAO).

```python
plt.figure(figsize=(12, 8))
sns.catplot(data=df, x='GESTACAO', y='APGAR5', hue='SEXO', kind='violin', split=True, height=8, aspect=1.5)
plt.title('Distribuição de APGAR5 por Sexo e Tempo de Gestação')
plt.show()
```

####  4) Monte uma tabela cruzada com APGAR5 nas colunas e GESTACAO nas linhas

```python
#tabela cruzada
tabela_cruzada = pd.crosstab(df['GESTACAO'], df['APGAR5'])

print(tabela_cruzada)
```

####  5) Da tabela cruzada do intem anterior, calcule os percentuais de linha. Ou seja, a distribuição da variável APGAR5 para cada classe de GESTACAO dado.

```python
#percentuais de linha
percentuais_linha = tabela_cruzada.div(tabela_cruzada.sum(axis=1), axis=0) * 100

print(percentuais_linha)
```

####  6) Os médicos costumam classificar parto prematuro quando a gestação tem menos que 37 semanas. Faça um gráfico de barras da tabela cruzada acima e conclua se você acha que este conjunto de dados corrobora ou não com a afirmação dos médicos.

```python
percentuais_linha.plot(kind='bar', stacked=True, figsize=(12, 8))
plt.title('Distribuição de APGAR5 por Grupo de Tempo de Gestação')
plt.xlabel('Tempo de Gestação')
plt.ylabel('Percentual')
plt.legend(title='APGAR5')
plt.show()
```

### grafico em boxplot

```python
# Criar uma coluna categorizada para o peso do bebê
df['PESO_cat'] = pd.cut(df['PESO'], bins=[0, 1500, 2500, 3500, 4500, 6000], labels=['Muito baixo', 'Baixo', 'Normal', 'Alto', 'Muito alto'])

plt.figure(figsize=(12, 8))
sns.boxplot(data=df, x='PESO_cat', y='APGAR5')
plt.title('Distribuição de APGAR5 por Peso do Bebê')
plt.xlabel('Peso do Bebê')
plt.ylabel('APGAR5')
plt.show()
```

### grafico em violino

```python
plt.figure(figsize=(12, 8))
sns.violinplot(data=df, x='PESO_cat', y='APGAR5')
plt.title('Distribuição de APGAR5 por Peso do Bebê (Violin Plot)')
plt.xlabel('Peso do Bebê')
plt.ylabel('APGAR5')
plt.show()
```
